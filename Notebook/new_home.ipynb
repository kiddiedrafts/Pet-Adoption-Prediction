```python
# loading the data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DL/5.deep_neural_networks/new_home/data/petfinder_train.csv')
data
```

```python
# splitting the dataset
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(data, test_size = 0.2)
```

```python
# creating a binary target variable based on AdoptionSpeed
train_data['Target'] = train_data['AdoptionSpeed'] < 4
train_data['Target']
```

```python
test_data['Target'] = test_data['AdoptionSpeed'] < 4
test_data['Target']
```

```python
train_data.drop(columns = ['AdoptionSpeed', 'Description'], inplace = True)
test_data.drop(columns = ['AdoptionSpeed', 'Description'], inplace = True)
```

```python
train_data.info()
```

```python
# checking the number of unique values in each specified column
columns = ["Type", "Breed1", "Gender", "Color1", "Color2", "MaturitySize", "FurLength", "Vaccinated", "Sterilized", "Health"]

for column in columns:
    print("column:", column, "unique_values:", len(train_data[column].unique()))
```

```python
# encoding categorical columns with LabelEncoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

train_data['MaturitySize'] = label_encoder.fit_transform(train_data['MaturitySize'])
test_data['MaturitySize'] = label_encoder.transform(test_data['MaturitySize'])

train_data['FurLength'] = label_encoder.fit_transform(train_data['FurLength'])
test_data['FurLength'] = label_encoder.transform(test_data['FurLength'])

train_data['Health'] = label_encoder.fit_transform(train_data['Health'])
test_data['Health'] = label_encoder.transform(test_data['Health'])
```

```python
# encoding 'Type' and 'Gender' columns with LabelEncoder
train_data['Type'] = label_encoder.fit_transform(train_data['Type'])
test_data['Type'] = label_encoder.transform(test_data['Type'])

train_data['Gender'] = label_encoder.fit_transform(train_data['Gender'])
test_data['Gender'] = label_encoder.transform(test_data['Gender'])
```

```python
# encoding categorical columns with BinaryEncoder
import category_encoders as ce

binary_encoder = ce.BinaryEncoder()

train_binary_Breed1 = binary_encoder.fit_transform(train_data['Breed1'])
test_binary_Breed1 = binary_encoder.transform(test_data['Breed1'])

train_binary_Color1 = binary_encoder.fit_transform(train_data['Color1'])
test_binary_Color1 = binary_encoder.transform(test_data['Color1'])

train_binary_Color2 = binary_encoder.fit_transform(train_data['Color2'])
test_binary_Color2 = binary_encoder.transform(test_data['Color2'])

train_binary_Vaccinated = binary_encoder.fit_transform(train_data['Vaccinated'])
test_binary_Vaccinated = binary_encoder.transform(test_data['Vaccinated'])

train_binary_Sterilized = binary_encoder.fit_transform(train_data['Sterilized'])
test_binary_Sterilized = binary_encoder.transform(test_data['Sterilized'])
```

```python
# adding the encoded columns and dropping the original ones from the train and test data
train_data = pd.concat([train_data, train_binary_Breed1, train_binary_Color1, train_binary_Color2, train_binary_Vaccinated, train_binary_Sterilized], axis=1)
columns = ['Breed1','Color1','Color2','Vaccinated','Sterilized']
train_data.drop(columns = columns, inplace = True)

test_data = pd.concat([test_data, test_binary_Breed1, test_binary_Color1, test_binary_Color2, test_binary_Vaccinated, test_binary_Sterilized], axis=1)
columns = ['Breed1','Color1','Color2','Vaccinated','Sterilized']
test_data.drop(columns = columns, inplace = True)

train_data.info()
```

```python
# standardizing the features (excluding the 'Target' column)
columns = train_data.columns.difference(['Target'])
for column in columns:
  mean = train_data[column].mean()
  std = train_data[column].std()
  train_data[column] = (train_data[column]- mean)/std
  test_data[column] = (test_data[column] - mean)/std

train_data
```

```python
# splitting the data into training, validation, and test sets
X_train, X_valid, y_train, y_valid = train_test_split(train_data.drop(columns = ['Target']), train_data['Target'], test_size = 0.1)

X_test = test_data.drop(columns = ['Target'])
y_test = test_data['Target']

print('Train examples:', len(X_train), len(y_train))
print('Validation examples:', len(X_valid), len(y_valid))
print('Test examples:',  len(test_data), len(y_test))
```

```python
# defining a deep neural network model with 3 hidden layers for binary classification
model = keras.models.Sequential()
model.add(keras.layers.Input(shape = (26,)))
model.add(keras.layers.Dense(128, activation = 'relu'))
model.add(keras.layers.Dense(64, activation = 'relu'))
model.add(keras.layers.Dense(32, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
# compiling the model
model.compile(optimizer = 'adam',
              loss = keras.losses.BinaryCrossentropy(from_logits=False),
              metrics = ['accuracy'])
```

```python
# training the model
epochs = 10
BATCH_SIZE = 128

history = model.fit(X_train, y_train,
                    batch_size=BATCH_SIZE,
                    epochs=epochs,
                    validation_data=(X_valid, y_valid))
```

```python
# predicting on the train and test sets
y_train_pred = (model.predict(X_train, batch_size = 128) > 0.5).astype(bool)
y_test_pred = (model.predict(X_test, batch_size = 128) > 0.5).astype(bool)
```

```python
# model evaluation
from sklearn.metrics import f1_score

f1_train = f1_score(y_train, y_train_pred, average='weighted')
f1_test = f1_score(y_test, y_test_pred, average='weighted')

print(f"Model F1 score on training set: {round(f1_train, 2)}%")
print(f"Model F1 score on test set: {round(f1_test, 3)}%")
```
